* CHIRP http://chg.geog.ucsb.edu/data/chirps/

1) Importing Modules

```python
import os
import ftplib
```

2) Creating output folder or setting it

```python
outFolder = "V:/Precipitation"
if not os.path.exists(outFolder): os.mkdir(outFolder)
```

3) Creating Year and Month subfolders

```python
years = [2017, 2018]

for year in years:
    dateFolder = outFolder + os.sep + str(year)
    if not os.path.exists(dateFolder): os.mkdir(dateFolder)
    print(dateFolder)
```

4) Setting URLs and logging in

```python
ftpURL = "ftp.chg.ucsb.edu"
ftp = ftplib.FTP(ftpURL)
ftp.login("anonymous","user@duke.edu")
print(ftp.welcome)
```

5) Downloading CHIRP files iterating through 2017 & 2018
* warning 50+GB of data

```python
for year in years:
    Dir = "/pub/org/chg/products/CHIRP/daily/{}/".format(year)
    print(Dir)
    ftp.cwd(Dir)
    Files = ftp.nlst()
    for file in Files:
        if file == "india":
            continue
        fileName = file
        outFilename = outFolder+ os.sep+ str(year)+ os.sep+ fileName
        print(outFilename)
        ftp.retrbinary("RETR " + fileName,open(outFilename,'wb').write)
```

```python
ftp.close()
```

# Playing with reading the files
* Basic Numbpy Visualization

```python
import arcpy
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
precip1 = arcpy.Raster('V:/Precipitation/2017/chirp.2017.01.01.tif')
```

```python
desc = arcpy.Describe(precip1)

extent = desc.extent
spatialReference = desc.spatialReference
name = desc.name
dataType = desc.dataType

print(extent, spatialReference, name, dataType)
```

```python
arrPrecip = arcpy.RasterToNumPyArray(precip1)
```

```python
#Create the canvas, setting the size to 12 x 8 units
plt.figure(figsize=(12,8))
#Add the DEM on the canvas, using the terrain colormap 
plt.imshow(arrPrecip,cmap='terrain')
#Display the figure in the notebook
plt.show()
```
